```python
from pymongo import MongoClient
import json
from collections import defaultdict
```

```python
client = MongoClient('mongodb://localhost:27017')
db = client.geonames
collection = db.all_names
```

```python
results = collection.aggregate([
        {
            '$match':{
                'feature_code':{
                    '$in':['PCLI','PCLX','PCLS','PCLD','PCLF','PCL','PCLH'] #['PCLD']
                }
            }
        },
        {
            '$group':
                {
                '_id':{
                    'ascii_name':'$ascii_name'
                },
                'alternate_names':{
                    '$push':'$alternate_name'
                }
            }
        },
        {
            '$project':{
                '_id':0,
                'ascii_name':'$_id.ascii_name',
                'alternate_names':'$alternate_names'
            }
        }
    ])
```

```python
independent_countries = defaultdict(list)
for result in results:
    independent_countries[result['ascii_name']] =  result['alternate_names']
print len(independent_countries)
sorted(independent_countries.keys())
```

```python
for i in independent_countries["Republic of Cote d'Ivoire"]:
    print unicode(i)
```
